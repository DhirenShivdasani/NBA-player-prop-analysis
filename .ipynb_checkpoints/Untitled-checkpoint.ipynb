```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
all_data = pd.read_csv('all_data.csv')
lines_data = pd.read_csv('test2.csv')

# Preprocess the data (modify column names and formats as needed)
# For instance, ensure that the 'PlayerName' in all_data matches 'Name' in lines_data

# Merge the datasets on common keys (e.g., PlayerName and GameDate)
# Assuming 'GAME_DATE' and 'PlayerName' can be used for merging
combined_data = pd.merge(all_data, lines_data, left_on='PlayerName', right_on='Name')

# Function to determine if a player went over or under the line
def over_under(row, stat_type):
    if row[stat_type] > row['Value']:
        return 'Over'
    else:
        return 'Under'

# Apply the function for Points, Rebounds, and Assists
combined_data['PTS_Result'] = combined_data.apply(lambda row: over_under(row, 'PTS'), axis=1)
combined_data['REB_Result'] = combined_data.apply(lambda row: over_under(row, 'REB'), axis=1)
combined_data['AST_Result'] = combined_data.apply(lambda row: over_under(row, 'AST'), axis=1)

# Analysis: Find pairs of players who often go over their line in the same game
# Create a dataframe to store pairs and their frequencies
pairs_frequency = pd.DataFrame(columns=['Player1', 'Player2', 'Frequency'])

# Iterate over games and players
for game_id in combined_data['Game_ID'].unique():
    game_data = combined_data[combined_data['Game_ID'] == game_id]
    players = game_data['PlayerName'].unique()
    
    for i in range(len(players)):
        for j in range(i+1, len(players)):
            player1 = players[i]
            player2 = players[j]
            if game_data[(game_data['PlayerName'] == player1) & (game_data['PTS_Result'] == 'Over')].any() and \
               game_data[(game_data['PlayerName'] == player2) & (game_data['PTS_Result'] == 'Over')].any():
                # Update frequency
                if ((pairs_frequency['Player1'] == player1) & (pairs_frequency['Player2'] == player2)).any():
                    idx = pairs_frequency[(pairs_frequency['Player1'] == player1) & (pairs_frequency['Player2'] == player2)].index[0]
                    pairs_frequency.at[idx, 'Frequency'] += 1
                else:
                    pairs_frequency = pairs_frequency.append({'Player1': player1, 'Player2': player2, 'Frequency': 1}, ignore_index=True)

# Sort pairs by frequency
pairs_frequency.sort_values(by='Frequency', ascending=False, inplace=True)

# Visualization (optional)
# Plotting the top 10 pairs
top_pairs = pairs_frequency.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_pairs, x='Frequency', y='Player1')
plt.title('Top Pairs Going Over the Line Together')
plt.show()
```

```python
# app = Flask(__name__)

def extract_teams_from_matchup(matchup):
    if '@' in matchup:
        team, opponent = matchup.split(' @ ')
    elif 'vs.' in matchup:
        team, opponent = matchup.split(' vs. ')
    else:
        raise ValueError("Invalid matchup format")
    return team.strip(), opponent.strip()

def calculate_home_field(data):
    for index, row in data.iterrows():
        player_team, _ = extract_teams_from_matchup(row['MATCHUP'])
        if row['MATCHUP'].startswith(player_team):
            data.at[index, 'HOME'] = 1
        else:
            data.at[index, 'HOME'] = 0

def calculate_average_stats(player_data, features):
    last_5_games = player_data.sort_values(by='GAME_DATE', ascending=False).head(5)
    average_stats = last_5_games[features].mean().to_dict()
    return average_stats

def calculate_combined_stats(player_data, opponent, features):
    games_against_opponent = player_data[player_data['MATCHUP'].str.contains(opponent)]
    opponent_stats = games_against_opponent[features].mean().to_dict() if not games_against_opponent.empty else {}
    general_stats = player_data[features].mean().to_dict()
    combined_stats = {feature: opponent_stats.get(feature, general_stats.get(feature)) for feature in features}
    return combined_stats


def model_evaluation(X_train, X_test, y_train, y_test):
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)  # Debug: Check probabilities


    return accuracy_score(y_test, y_pred), precision_score(y_test, y_pred), recall_score(y_test, y_pred), f1_score(y_test, y_pred), model

# @app.route('/', methods=['GET'])
def matchups_form():
    return render_template('predictions_form.html')

def integrate_injury_data(performance_data, injury_data):
    # Mark injured players as unavailable
    for index, injury_row in injury_data.iterrows():
        player_name = injury_row['Player']
        player_team = injury_row['Team']
        
        # Find the latest game for the player's team
        latest_game = performance_data[(performance_data['PlayerName'] == player_name) & 
                                       (performance_data['MATCHUP'].str.contains(player_team))]['GAME_DATE'].max()
        
        if latest_game:
            # Mark the player's stats for the latest game as unavailable
            performance_data.loc[(performance_data['PlayerName'] == player_name) & 
                                 (performance_data['GAME_DATE'] == latest_game), 'Available'] = False
        else:
            # If no latest game found, mark all games as unavailable
            performance_data.loc[performance_data['PlayerName'] == player_name, 'Available'] = False

    # Adjust team stats/player averages considering injuries
    # This part of the code depends on how you want to adjust the stats.
    
    return performance_data


# @app.route('/predictions', methods=['POST'])
def predictions():
    data = pd.read_csv('all_data.csv')
    injury_data = pd.read_csv('injury_data.csv')
    data['GAME_DATE'] = pd.to_datetime(data['GAME_DATE'], format='%b %d, %Y')

    current_date = datetime.now().strftime('%a, %b %d, %Y')
    injury_data_filtered = injury_data[(injury_data['Update'] == current_date) | 
                                   (injury_data['Details'].str.contains("Out"))]
    print(injury_data_filtered)

    data = integrate_injury_data(data, injury_data)


    calculate_home_field(data)

    lines_data = pd.read_csv('test2.csv')
    prop_to_stat = {'Points': 'PTS', 'Rebounds': 'REB', 'Assists': 'AST'}

    # matchups_input = request.form['matchups']
    # matchups_lines = matchups_input.split('\n')

    predictions = {stat: [] for stat in ['PTS', 'REB', 'AST']}
    min_accuracy = 0.7

    # for line in matchups_lines:
    #     parts = line.split(' ')
    #     if len(parts) == 3:
    #         team1, team2, home_indicator = parts
    #         team1 = team1.strip()
    #         team2 = team2.strip()
    #         home_indicator = int(home_indicator.strip())

            
    relevant_players = data[(data['MATCHUP'].str.contains('DEN')) | (data['MATCHUP'].str.contains('GSW'))]['PlayerName'].unique()


    for player in relevant_players:
        player_data = data[data['PlayerName'] == player]
        player_lines = lines_data[lines_data['Name'] == player]

        for _, row in player_lines.iterrows():
            player_line_value = row['Value']
            stat_type = prop_to_stat.get(row['Prop'], None)

            if len(player_data) >= 5 and stat_type:
                X = player_data[['HOME','AST', 'STL', 'REB', 'TOV', 'FG3M', 'FG3A', 'BLK', 'FGA', 'FGM', 'FTA', 'FTM', 'PLUS_MINUS']]
                y = player_data[stat_type].apply(lambda x: 1 if x > player_line_value else 0)
                scaler = StandardScaler()
                X_scaled = scaler.fit_transform(X)
                X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.4, random_state=42)
                
                accuracy, precision, recall, f1, model = model_evaluation(X_train, X_test, y_train, y_test)
                
                if accuracy >= min_accuracy:
                    average_stats = calculate_average_stats(player_data, ['AST', 'STL', 'REB', 'TOV', 'FG3M', 'FG3A', 'BLK', 'FGA', 'FGM', 'FTA', 'FTM', 'PLUS_MINUS'])
                    X_avg = pd.DataFrame([average_stats])
                    feature_order = ['HOME'] + list(average_stats.keys())
                    X_avg['HOME'] = 1 
                    X_avg = X_avg[feature_order]
                    X_avg_scaled = scaler.transform(X_avg)
                    prediction = model.predict(X_avg_scaled)[0]
                    prob_over = model.predict_proba(X_avg_scaled)[0][1]
                    prediction_text = 'over' if prediction == 1 else 'under'
                    confidence_distance = abs(prob_over - 0.5)
                    predictions[stat_type].append((player, 'DEN' if player in 'DEN' else 'GSW', player_line_value, prediction_text, round(prob_over, 2), round(confidence_distance, 2)))

    for stat in predictions:
        predictions[stat].sort(key=lambda x: x[5], reverse=True)
    return predictions
    # return render_template('predictions.html', predictions=predictions)

```

```python
def adjust_stats_based_on_injuries(data, injury_data):
    adjusted_data = data.copy()

    # Ensure 'PlayerName' column exists
    if 'PlayerName' not in data.columns:
        raise KeyError("Column 'PlayerName' does not exist in the provided DataFrame.")
    
    for injured_player in injury_data[injury_data['Details'].str.contains("Out")]['Player'].unique():
        games_with_player = data[data['PlayerName'] == injured_player]
    
        if not games_with_player.empty:
            player_team = games_with_player['MATCHUP'].iloc[0].split(' ')[0]
            # Further processing
        else:
            print(f"No games found for injured player: {injured_player}")
            # Handle cases where no games are found for the injured player
        
        # Skip if no games found
        if games_with_player.size == 0:
            continue
    
        # Find games where the injured player was absent
        games_without_player = data[(~data['Player_ID']) & (data['MATCHUP'].str.contains(player_team))]
    
        # Ensure there are games without the player
        if not games_without_player.empty:
            continue
        else:
            print(f"No games found for injured player: {injured_player}")
    
        numeric_cols = all_data.select_dtypes(include=['number']).columns
        avg_stats_with_player = data[data['Game_ID'].isin(games_with_player)][numeric_cols].groupby('PlayerName').mean()
        avg_stats_without_player = games_without_player[numeric_cols].groupby('PlayerName').mean()

        # Adjust stats for players based on the injured player's absence
        for player, stats in avg_stats_without_player.iterrows():
            if player in avg_stats_with_player.index:
                performance_change = stats - avg_stats_with_player.loc[player]
                adjustment_factor = .1
                adjusted_stats = performance_change * adjustment_factor
                adjusted_data.loc[adjusted_data['PlayerName'] == player, numeric_cols] += adjusted_stats
    return adjusted_data
```

```python
def calculate_adjustment_factor(data, games_with_player, games_without_player, key_stats):
    # Filter data for key statistics
    data_with_player = data[data['Game_ID'].isin(games_with_player)][key_stats]
    data_without_player = data[data['Game_ID'].isin(games_without_player)][key_stats]

    # Calculate mean values
    mean_with_player = data_with_player.mean()
    mean_without_player = data_without_player.mean()

    # Calculate percentage change for each key stat
    percentage_change = (mean_without_player - mean_with_player) / mean_with_player

    # Optional: Check for statistical significance here

    # Average the percentage changes across all key stats to get a single adjustment factor
    adjustment_factor = percentage_change.mean()

    return adjustment_factor
```
